```python
import os
import json
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
os.getcwd()
# wczytaj plik JSON
with open('coco.json') as f:
    data = json.load(f)
```

```python
# cieka do folderu, w ktrym znajduje si plik JSON z danymi
data_folder = os.getcwd()

# cieka do folderu, w ktrym znajduj si obrazy
image_folder = os.path.join(data_folder, 'images\\train')

# cieka do folderu, w ktrym maj by zapisane wycinki z obrazw
cropped_folder = os.path.join(data_folder, 'cell_img')
```

```python
# Utworzenie sownika folderw dla kadej kategorii
category_folders = {}
for i in range(len(data['categories'])):
    objects = data['categories'][i]['name']
    if objects not in category_folders:
        folder_path = os.path.join(cropped_folder, objects)
        os.makedirs(folder_path, exist_ok=True)
        category_folders[objects] = folder_path
```

```python


for i in range(len(data['images'])):
    image_id = data['images'][i]['id']
    pathname = data['images'][i]['file_name'][8:]
    # Utworzenie penej cieki do pliku z obrazem
    image_path = os.path.join(image_folder ,pathname)
    # Wczytanie obrazu za pomoc biblioteki CV2
    image = cv2.imread(image_path)
    annotations = data['annotations']
    for j in range(len(annotations)):
        if annotations[j]['image_id'] == image_id:
            x,y,w,h = annotations[j]['bbox']
            cell_id = annotations[j]['category_id']
#         # Wycicie fragmentu obrazu
            cropped_image = image[y:y+h, x:x+w]# cropped image
#         # Zapisanie wycinka z obrazu na dysku
            cat = data['categories']
            for k in range(len(cat)):
                if cat[k]['id'] == cell_id:
                    category = cat[k]['name']#type of cell
                    folder_path = category_folders[category]#create directory to save
                    cv2.imwrite(os.path.join(data_folder, folder_path , str(j)+'_'+pathname), cropped_image)#pathname and mini image
```
